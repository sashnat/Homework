```python
import os
import urllib.request
import zipfile
import pandas as pd
```

```python
path = r'C:\\Users\E277460\PycharmProjects\stepic_python_practice\\2_4 from web'
file_path = path + '\\rogaikopyta_scr.zip'
unzip_path = path + '\\unzip'
#Скачивание файла в указанное место с указанием нового имени файла
urllib.request.urlretrieve("https://stepik.org/media/attachments/lesson/245299/rogaikopyta.zip", file_path)
#Распаковка файла
with zipfile.ZipFile(file_path, 'r') as zip_ref:
    zip_ref.extractall(unzip_path)
```

```python
d = {}
for file in os.listdir(unzip_path):
    df_n = pd.read_excel(unzip_path + "\\" + file)
    key, val = df_n.loc[0].values[1], df_n.loc[0].values[3]
    d[key] = val

for key, val in sorted(d.items(), key = lambda x: x[0]):
    print(key, int(val))
```
